## Throw

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
g = -9.81
timestep = 0.001
```

#### Initial conditions

```python
x_0 = 0
y_0 = 0.005
angle_0_deg = 30
v_0 = 60
```

#### Initial velocity components

```python
angle_0 = np.radians(angle_0_deg)
v_0_x = v_0 * np.cos(angle_0)
v_0_y = v_0 * np.sin(angle_0)
```

#### Simulation

```python
x = x_0
y = y_0
x_log = [x_0]
y_log = [y_0]
time_log = [0]
v_x = v_0_x
v_y = v_0_y
while (y > 0):
    x = x + v_x * timestep
    y = y + v_y * timestep
    v_y = v_y + g * timestep
    time_log.append(time_log[-1] + timestep)
    x_log.append(x)
    y_log.append(y)
```

#### Plots

```python
y_axis = max(y_log)
x_axis = max(x_log)
if y_axis * 3 > x_axis:
    y_lim = y_axis
    x_lim = y_lim * 3
else:
    x_lim = x_axis
    y_lim = x_lim / 3
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(15,5))
ax.set_title('Trajectory')
ax.set_xlabel('Distance [m]')
ax.set_ylabel('Height [m]')
ax.grid(linestyle='--')
ax.plot(x_log,y_log)
ax.axis((0, x_lim, 0, y_lim))
plt.show()
```
